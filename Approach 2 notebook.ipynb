```python
#importing data, important packages and cleaning the data
#Selected SKU = Cucumber (Indian)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/home/Akai/Documents/Documents/Ninjacart/cleaned_data.csv")
spe_data = data[data.SKUName == 'Cucumber (Indian)'].reset_index(drop=True)
spe_data = spe_data.drop(labels = ['FulfilledQty', 'ReturnQty', 'StockOutAfterTime', 'OSCustomerCount', 'Late1Hour', 'MissedDemand', 'OnTimeDelivery'], axis=1)

for i in range(len(spe_data)):
    if spe_data.OrderedQty[i]<np.percentile(spe_data.OrderedQty, 5):
        spe_data.OrderedQty[i]=np.percentile(spe_data.OrderedQty, 5)
    elif spe_data.OrderedQty[i]>np.percentile(spe_data.OrderedQty, 95):
        spe_data.OrderedQty[i]=np.percentile(spe_data.OrderedQty, 95)
    else:
        count=0
    if spe_data.CustomerCount[i]<np.percentile(spe_data.CustomerCount, 5):
        spe_data.CustomerCount[i]=np.percentile(spe_data.CustomerCount, 5)
    elif spe_data.CustomerCount[i]>np.percentile(spe_data.CustomerCount, 95):
        spe_data.CustomerCount[i]=np.percentile(spe_data.CustomerCount, 95)
    else:
        riario=0
    if spe_data.BilledQty[i]<np.percentile(spe_data.BilledQty, 5):
        spe_data.BilledQty[i]=np.percentile(spe_data.BilledQty, 5)
    elif spe_data.BilledQty[i]>np.percentile(spe_data.BilledQty, 95):
        spe_data.BilledQty[i]=np.percentile(spe_data.BilledQty, 95)
    else:
        vernon=0
    if spe_data.AvgSP[i]<np.percentile(spe_data.AvgSP, 5):
        spe_data.AvgSP[i]=np.percentile(spe_data.AvgSP, 5)
    elif spe_data.AvgSP[i]>np.percentile(spe_data.AvgSP, 95):
        spe_data.AvgSP[i]=np.percentile(spe_data.AvgSP, 95)
    else:
        anite=0
    if spe_data.Wholesale[i]<np.percentile(spe_data.Wholesale, 5):
        spe_data.Wholesale[i]=np.percentile(spe_data.Wholesale, 5)
    elif spe_data.Wholesale[i]>np.percentile(spe_data.Wholesale, 95):
        spe_data.Wholesale[i]=np.percentile(spe_data.Wholesale, 95)
    else:
        eziquel=0
    if spe_data.RetailPrice[i]<np.percentile(spe_data.RetailPrice, 5):
        spe_data.RetailPrice[i]=np.percentile(spe_data.RetailPrice, 5)
    elif spe_data.RetailPrice[i]>np.percentile(spe_data.RetailPrice, 95):
        spe_data.RetailPrice[i]=np.percentile(spe_data.RetailPrice, 95)
    else:
        lavezzi=0
    if spe_data.FinalGRN[i]<np.percentile(spe_data.FinalGRN, 5):
        spe_data.FinalGRN[i]=np.percentile(spe_data.FinalGRN, 5)
    elif spe_data.FinalGRN[i]>np.percentile(spe_data.FinalGRN, 95):
        spe_data.FinalGRN[i]=np.percentile(spe_data.FinalGRN, 95)
    else:
        ricardo=0
    if spe_data.TotalGTOrders[i]<np.percentile(spe_data.TotalGTOrders, 5):
        spe_data.TotalGTOrders[i]=np.percentile(spe_data.TotalGTOrders, 5)
    elif spe_data.TotalGTOrders[i]>np.percentile(spe_data.TotalGTOrders, 95):
        spe_data.TotalGTOrders[i]=np.percentile(spe_data.TotalGTOrders, 95)
    else:
        quaresma=0
    if spe_data.ActualDemand[i]<np.percentile(spe_data.ActualDemand, 5):
        spe_data.ActualDemand[i]=np.percentile(spe_data.ActualDemand, 5)
    elif spe_data.ActualDemand[i]>np.percentile(spe_data.ActualDemand, 95):
        spe_data.ActualDemand[i]=np.percentile(spe_data.ActualDemand, 95)
    else:
        count=0
        
spe_data = spe_data.ix[8:].reset_index(drop=True)
```

```python
#setting input and output variables

onion_data = data[data.SKUName == 'Onion'].reset_index(drop=True)
onion_data = onion_data.ix[8:].reset_index(drop=True)
dummy_variable1 = onion_data.AvgSP
tomato_data = data[data.SKUName == 'Tomato (local)'].reset_index(drop=True)
tomato_data = tomato_data.ix[8:].reset_index(drop=True)
dummy_variable2 = tomato_data.AvgSP

input_variables = pd.DataFrame({'Date':spe_data.DeliveryDate,'CustomerCount':spe_data.CustomerCount, 'AvgSP':spe_data.AvgSP, 'Wholesale':spe_data.Wholesale, 'RetailPrice':spe_data.RetailPrice, 'FinalGRN':spe_data.FinalGRN, 'TotalGTOrders':spe_data.TotalGTOrders, 'OP':dummy_variable1})
for i in range(len(input_variables)):
    input_variables.Date[i] = input_variables.Date[i][3:6] + input_variables.Date[i][:3] + '2017'
input_variables['Date'] = pd.to_datetime(input_variables['Date'])
input_variables.index = input_variables['Date']
input_variables = input_variables.drop(labels = ['Date'], axis=1)
input_variables = input_variables.drop(input_variables.index[len(input_variables)-1])

output_variable = pd.DataFrame({'Date':spe_data.DeliveryDate, 'ActualDemand':spe_data.ActualDemand})
for i in range(len(output_variable)):
    output_variable.Date[i] = output_variable.Date[i][3:6] + output_variable.Date[i][:3] + '2017'
output_variable['Date'] = pd.to_datetime(output_variable['Date'])
output_variable.index = output_variable['Date']
output_variable = output_variable.drop(labels = ['Date'], axis=1)
output_variable = output_variable.drop(output_variable.index[len(output_variable)-1])
```

```python
#Creating a dataframe to predict the CustomerCount and the total GT orders for the days requiring order quantity prediction 

from sklearn.linear_model import Ridge

cc_determining_df = pd.DataFrame({'Date':spe_data.DeliveryDate, 'AvgSP':spe_data.AvgSP, 'OP':dummy_variable1})
for i in range(len(cc_determining_df)):
    cc_determining_df.Date[i] = cc_determining_df.Date[i][3:6] + cc_determining_df.Date[i][:3] + '2017'
cc_determining_df['Date'] = pd.to_datetime(cc_determining_df['Date'])
cc_determining_df.index = cc_determining_df['Date']
cc_determining_df = cc_determining_df.drop(labels = ['Date'], axis=1)
cc_determining_df = cc_determining_df.drop(cc_determining_df.index[len(cc_determining_df)-1])
```

```python
#models to predict CustomerCount and TotalGTOrders
#pipeline first converts the features to polynomial features and then performs regression under the set parameters

from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures

model_cc = make_pipeline(PolynomialFeatures(degree=3), Ridge(alpha=18.0))
for_cc = model_cc.fit(cc_determining_df[0:58], input_variables.CustomerCount[0:58])
predictions_cc_train =for_cc.predict(cc_determining_df[0:58])
predictions_cc = for_cc.predict(cc_determining_df[58:])

model_gt = make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=21.0))
for_gt = model_gt.fit(cc_determining_df[0:58], input_variables.TotalGTOrders[0:58])
predictions_gt_train = for_gt.predict(cc_determining_df[0:58])
predictions_gt = for_gt.predict(cc_determining_df[58:])
```

```python
#calculating prediction accuracy

sse_cc = 0
for i in range(len(input_variables.CustomerCount[58:])):
    sse_cc+=(input_variables.CustomerCount[i+58] - predictions_cc[i] )**2

sst_cc = 0
for i in range(len(input_variables.CustomerCount[58:])):
    sst_cc=(input_variables.CustomerCount[i+58] - np.mean(input_variables.CustomerCount[0:58]))**2

residuals_cc = []
for i in range(len(input_variables.CustomerCount[58:])):
    residuals_cc.append(input_variables.CustomerCount[i+58] - predictions_cc[i])
    
r_sq_cc = 1-(sse_cc/sst_cc)
    
rmse_cc = (sse_cc/len(input_variables.CustomerCount[58:]))**0.5
print(rmse_cc, np.mean(residuals_cc), r_sq_cc)

sse_gt = 0
for i in range(len(input_variables.TotalGTOrders[58:])):
    sse_gt+=(input_variables.TotalGTOrders[i+58] - predictions_gt[i])**2

sst_gt = 0
for i in range(len(input_variables.TotalGTOrders[58:])):
    sst_gt=(input_variables.TotalGTOrders[i+58] - np.mean(input_variables.TotalGTOrders[0:58]))**2

residuals_gt = []
for i in range(len(input_variables.TotalGTOrders[58:])):
    residuals_gt.append(input_variables.TotalGTOrders[i+58] - predictions_gt[i])
    
r_sq_gt = 1-(sse_gt/sst_gt)
    
rmse_gt = (sse_gt/len(input_variables.TotalGTOrders[58:]))**0.5
print(rmse_gt, np.mean(residuals_gt), r_sq_gt)

predicted_values_cc = []
for i in range(len(predictions_cc)):
    predicted_values_cc.append(predictions_cc[i])
    
actual_values_cc = []
for i in range(len(predictions_cc)):
    actual_values_cc.append(input_variables.CustomerCount[i+58])
    
predicted_values_gt = []
for i in range(len(predictions_gt)):
    predicted_values_gt.append(predictions_gt[i])
    
actual_values_gt = []
for i in range(len(predictions_gt)):
    actual_values_gt.append(input_variables.TotalGTOrders[i+58])
    
predicted_values_cc_train = []
for i in range(len(predictions_cc_train)):
    predicted_values_cc_train.append(predictions_cc_train[i])
    
    
predicted_values_gt_train = []
for i in range(len(predictions_gt_train)):
    predicted_values_gt_train.append(predictions_gt_train[i])
```

```python
#plotting the predicted input values

p1 = plt.plot(predicted_values_cc, color = 'red', label = 'predicted values')
p2 = plt.plot(actual_values_cc, color = 'blue', label = 'actual values')
plt.xlabel('Prediction Days')
plt.ylabel('CustomerCount')
plt.legend()
plt.show(block=False)
```
